# 10절. 데이터 구조 변경
- 와이드포맷 <-> 롱포맷
    * 와이드 포맷 (wide format): 여러 변수(열)가 가로로 나열된 형식 (엑셀처럼)
    * 롱 포맷 (long format): 동일한 열 안에 변수 이름과 값이 세로로 나열된 형식
- melt()를 쓰면 와이드 → 롱 포맷으로 변환할 수 있습니다.
- 판다스 공식 문서의 melt :
https://pandas.pydata.org/docs/user_guide/reshaping.html#reshaping by melt
- Tidy Dayt(7page) : https://vita.had.co.nz/papers/tidy-data.pdf

```python
# melt함수 사용법 : 기본문법
pd.melt(
    frame,                 # 변환할 데이터프레임
    id_vars=None,          # 고정할 열 (식별자 역할)
    value_vars=None,       # 녹여서 변수로 만들 열 (지정하지 않으면 나머지 전부)
    var_name='variable',   # 새로 생길 '변수명' 열 이름
    value_name='value'     # 새로 생길 '값' 열 이름
)
```

```python
# melt함수 예제(원본 : 와이드 포멧)
import pandas as pd

df = pd.DataFrame({
    '이름': ['철수', '영희'],
    '국어': [90, 95],
    '수학': [85, 100]
})
```

```python
# ★ 2차원 데이터 프레임의 apply 이용하는 방법
```

```python
def split_date(df):
    df = df.copy()
    df['연도'] = int(df['연도월'].split('년')[0])
    df['월'] = int(df['연도월'].split('년')[1][:-1])
    return df
```

```python
split_date(new_df.iloc[0])
```

```python
new_df = new_df.apply(split_date, axis=1)
new_df.sample()
```

```python
def year_month_append(df):
    'df를 받아 연도와 월 추가후 return'
    df = df.copy() #깊은 복사
    df['연도'] = int(df['연도월'].split('년')[0])
    df['월'] = int(df['연도월'].split('년')[1].replace('월',''))
    return df
```

```python
year_month_append(new_df.iloc[0])
```

```python
new_df = new_df.apply(year_month_append, axis=1)
new_df.head()
```

# 12절. 일괄변경하기(결측치, 특정값)
## 12.1 fillna
- 대부분 결측치는 drop하거나, 평균값대체, 중위수대체, 그룹별 평균값, 그룹별 중위수, 이전값, 이후값 등으로 대체

```python
df = pd.DataFrame([[np.nan, 2, np.nan, 0],
                  [3,       4, np.nan, 1],
                  [np.nan,  3, np.nan, 5],
                  [np.nan,  3, np.nan, 4]],
                 columns=[list('ABCD')])
df
```

```python
# 모든 결측치를 0 대체
df.fillna(0)
```

```python
#결측치를 이전행의 값으로 대체
df.fillna(method='ffill')
```

```python
# 결측치를 다음행의 값으로 대체
df.fillna(method='bfill')
```

```python
# 결측치를 열마다 특정값으로 대체
values = {'A':85, 'B':80, 'C':75, 'D':90}
df.fillna(value=values)
```

```python
# 결측치를 열별 평균값으로 대체
df.mean(axis=0)
df.fillna(value=df.mean(axis=0))
```

```python
# 결측치 대체를 2번째 결측치까지만 대체
print(values)
df.fillna(value=values, limit=2)
```
