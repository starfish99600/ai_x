# 10절. 데이터 구조 변경
- 와이드포맷 <-> 롱포맷
    * 와이드 포맷 (wide format): 여러 변수(열)가 가로로 나열된 형식 (엑셀처럼)
    * 롱 포맷 (long format): 동일한 열 안에 변수 이름과 값이 세로로 나열된 형식
- melt()를 쓰면 와이드 → 롱 포맷으로 변환할 수 있습니다.
- 판다스 공식 문서의 melt :
https://pandas.pydata.org/docs/user_guide/reshaping.html#reshaping by melt
- Tidy Dayt(7page) : https://vita.had.co.nz/papers/tidy-data.pdf

```python
# melt함수 사용법 : 기본문법
pd.melt(
    frame,                 # 변환할 데이터프레임
    id_vars=None,          # 고정할 열 (식별자 역할)
    value_vars=None,       # 녹여서 변수로 만들 열 (지정하지 않으면 나머지 전부)
    var_name='variable',   # 새로 생길 '변수명' 열 이름
    value_name='value'     # 새로 생길 '값' 열 이름
)
```

```python
# melt함수 예제(원본 : 와이드 포멧)
import pandas as pd

df = pd.DataFrame({
    '이름': ['철수', '영희'],
    '국어': [90, 95],
    '수학': [85, 100]
})
```

```python
# ★ 2차원 데이터 프레임의 apply 이용하는 방법
```

```python
def split_date(df):
    df = df.copy()
    df['연도'] = int(df['연도월'].split('년')[0])
    df['월'] = int(df['연도월'].split('년')[1][:-1])
    return df
```

```python
split_date(new_df.iloc[0])
```

```python
new_df = new_df.apply(split_date, axis=1)
new_df.sample()
```

```python
def year_month_append(df):
    'df를 받아 연도와 월 추가후 return'
    df = df.copy() #깊은 복사
    df['연도'] = int(df['연도월'].split('년')[0])
    df['월'] = int(df['연도월'].split('년')[1].replace('월',''))
    return df
```

```python
year_month_append(new_df.iloc[0])
```

```python
new_df = new_df.apply(year_month_append, axis=1)
new_df.head()
```
