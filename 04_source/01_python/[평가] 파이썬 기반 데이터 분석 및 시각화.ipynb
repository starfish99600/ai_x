# 문제 2

```python
# 데이터 로드 및 변수저장
df_last = pd.read_csv('C:/AI_x/01_download/shareData/부동산_250213/주택도시보증공사_전국 신규 민간아파트 분양가격 동향_20241231.csv', encoding='cp949')
df_last.head()
```

```python
def commadrop2(price):
    if pd.isna(price):
        return np.nan
    elif price.strip() == '':
        return np.nan
    elif price.find(',') != -1:
        return int(price.replace(',',''))
    else:
        return int(price)

commadrop2(df_last.loc[0, '분양가격(제곱미터)'])
commadrop2(df_last.loc[1, '분양가격(제곱미터)']) 
commadrop2(df_last.loc[28, '분양가격(제곱미터)']) 
commadrop2(df_last.loc[368, '분양가격(제곱미터)'])
```

```python
df_last['제곱미터당분양가격'] = df_last['분양가격(제곱미터)'].map(commadrop2)
df_last.tail()
```

```python
df_last['평당분양가격'] = df_last['제곱미터당분양가격']*3.3
df_last.head()
```

```python
df_last.drop(['제곱미터당분양가격'], axis=1, inplace=True)
```

```python
df_last.drop(['분양가격(제곱미터)'], axis=1, inplace=True)
```
